## Description

- Game has 2 players: `Player 1` and `Player 2`.

- `Player 1` goes first.

- `Player 1` gets a choice of choosing `X` or `O` as his symbol.

- After each move of a player, output the board.

- To mark positions on the mark, use the following system:

```
1 2 3
4 5 6
7 8 9
```

- Code uses just a bunch of functions.

````python
# define the board
board = [[' ' for j in range(3)] for i in range(3)]
print(board)
````

````python
def show_board(board):
    '''
    displays the current state of the board
    assumes 'board' is a list of 3*3 size
    '''
    n, m = 3, 3
    for i in range(n):
        for j in range(m):
            end_char = ' | '
            if j == m - 1:
                end_char = '\n'
            print(board[i][j], end = end_char)
        if i != n - 1:
            print('-' * (n * m))

show_board(board)
````

````python
from math import ceil

def map_position_to_grid(pos):
    '''
    maps a position(1-9) to a grid cell
    '''
    row = ceil(pos / 3)
    col = pos % 3
    if col == 0:
        col = 3
    return (row - 1, col - 1)

for i in range(1, 10):
    print(i, '->', map_position_to_grid(i))
````

````python
def accept_player_input(player, board):
    '''
    accepts an input position from a player
    runs a loop until input is valid
    '''
    while True:
        pos = input(f'Player {player}, enter a position between 1 - 9: ')
        if pos.isdigit() and pos != '0':
            pos = int(pos)
            row, col = map_position_to_grid(pos)
            if board[row][col] == ' ':
                return pos
            else:
                print('The position is occupied')
        else:
            print('Wrong input, please enter a position between 1 to 9!')

# accept_player_input(1, board)
````

````python
def update_board(player, player_symbol, board, pos):
    '''
    updates the board as per the player's move
    '''
    row, col = map_position_to_grid(pos)
    board[row][col] = player_symbol
    print('Move completed, heres the updated board:')
    show_board(board)
    
# update_board(1, 'X', board, 3)
# update_board(2, 'O', board, 1)
````

````python
def check_if_game_is_complete(board):
    '''
    check if the game is complete based on current state of the board
    '''
    symbols = ['X', 'O']
    for symbol in symbols:
        
        # check rows
        for i in range(3):
            if (board[i] == [symbol for j in range(3)]):
                return True
        
        # check cols
        for j in range(3):
            cols = True
            for i in range(3):
                cols = cols and (board[i][j] == symbol)
            if cols:
                return True
        
        # check diagonals
        diag = ([board[0][0], board[1][1], board[2][2]] == [symbol for k in range(3)])
        diag = diag or ([board[0][2], board[1][1], board[2][0]] == [symbol for k in range(3)])
        if diag:
            return True
    
    return False

# print(check_if_game_is_complete(board))
# show_board(board)
# board[0][2] = board[1][2] = board[2][2] = 'X'
# show_board(board)
# print(check_if_game_is_complete(board))
````

````python
def play_game(board, player_1_symbol):
    '''
    contains the actual game playing logic
    '''
    player_2_symbol = 'O'
    if player_1_symbol == 'O':
        player_2_symbol = 'X'
    players = [(1, player_1_symbol), (2, player_2_symbol)]
    game_over = False
    turn = 1
    
    while not game_over:
    
        if turn == 1:
            turn = 0
        else:
            turn = 1
        player_number, player_symbol = players[turn]
        
        print(f'**Player {player_number}\'s turn**')
        pos = accept_player_input(player_number, board)
        update_board(player_number, player_symbol, board, pos)
        
        draw = (board[0].count(' ') == 0 and board[1].count(' ') == 0 and board[2].count(' ') == 0)
        
        if check_if_game_is_complete(board):
            print(f'ðŸ¤©ðŸŽ‰ðŸŽŠPlayer {player_number} won!!! Game overðŸŽŠðŸŽ‰ðŸ¤©')
            game_over = True
        elif draw:
            game_over = True
            print('ðŸ’¥DRAW!!! Better luck next time.ðŸ’¥')
        else:
            print('')
````
